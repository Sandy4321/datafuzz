```python
from datafuzz.generators import DatasetGenerator
from datafuzz import Duplicator
from datetime import datetime, timedelta
from pprint import pprint
```

```python
generator = DatasetGenerator({
    'num_rows': 1000,
    'schema': {'machine_id': range(1,10),
               'operator_name': 'faker.name',
               'warehouse': ['main', 'secondary', 'outsourced'],
               'malfunction': range(1),
              },
    'start_time': datetime.now() - timedelta(days=5),
    'output': 'dataset',
    'increments': 'hours',
})
```

```python
generator.generate()
```

```python
for record in generator.records[:2]:
    pprint(record)
```

```python
dataset = generator.to_output()
```
